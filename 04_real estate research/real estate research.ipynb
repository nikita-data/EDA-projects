**Описание проекта**

В рамках данного проекта мы подготовим исследование объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктов за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Наша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 


**Предоставленные файлы**

#### Замена пропусков по столбцу ceiling_height

Заполнение пропусков исходя из зависимости  расстояния до центра города.

Заполним пропуски, если расстояние до центра города меньше 5000 м.

```python
data.loc[(data['city_centers_nearest'] < 5000) & (data['ceiling_height'].isna()),'ceiling_height'] = data.loc[data['city_centers_nearest'] < 5000,'ceiling_height'].median()
```

Заполним пропуски, если расстояние до центра города меньше 15000 м.

```python
data.loc[(data['city_centers_nearest'] < 15000) & (data['ceiling_height'].isna()),'ceiling_height'] = data.loc[data['city_centers_nearest'] < 15000,'ceiling_height'].median()
```

Заполним пропуски, если расстояние до центра города меньше 30000 м.

Заполним пропуски, если расстояние до центра города меньше 66000 м.

```python
data.loc[(data['city_centers_nearest'] < 66000) & (data['ceiling_height'].isna()),'ceiling_height'] = data.loc[data['city_centers_nearest'] < 66000,'ceiling_height'].median()
```

Заполнение оставшихся пропусков исходя из зависимости кол-ва этажей в доме.

```python
unique_floors_total = data['floors_total'].unique()
for index in unique_floors_total:    
    data.loc[(data['floors_total'] == index) & (data['ceiling_height'].isna()),'ceiling_height'] =data.loc[data['floors_total'] == index,'ceiling_height'].median()
```

Произведём расчёт цены квадратного метра по каждому объекту.Стоимость квадратного метра это отношение общей стоимости квартиры к её полной площади

```python
data['month'] = data['first_day_exposition'].dt.month
```

### Год публикации

Добавим столбец с указанием года публикации объявления

```python
data['year'] = data['first_day_exposition'].dt.year
```

### Тип этажа квартиры

Добавим столбец с указанием типа этажа квартиры, где значения — «первый», «последний», «другой».

```python
def type_floor(row):
    if row['floor'] == 1:
        return 'первый'
    if row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
    
```

```python
data['floor_name'] = data.apply(type_floor,axis=1)
```
