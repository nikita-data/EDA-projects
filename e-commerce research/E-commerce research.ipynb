**Описание проекта**

В рамках данного исследования мы произведём анализ интернет-магазина «Стримчик», который продаёт по всему миру компьютерные игры.  
Из открытых источников нам доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

**Предоставленный файл**

**Дополнительная информация о данных:**
- В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

### Анализ платформ - лидеров по продажам, потенциальной прибыльности.

Для определения платформ, лидирующих по продажам и их тренда (растут или падают), а также выбора потенциально прибыльных платформ, мы можем проанализировать данные актуального периода, который мы определили ранее (с 2013 по 2016 годы).

```python
# Группировка по платформам и подсчет суммарных продаж за актуальный период
platform_sales_actual = data_actual.pivot_table(index = 'platform',values='total_sales',aggfunc='sum').reset_index()
platform_sales_actual.columns = ['Название платформы','Суммарные продажи']

# Сортировка платформ по суммарным продажам
platform_sales_sorted = platform_sales_actual.sort_values(by='Суммарные продажи', ascending=False)
```

На конец 2016 года лидируют по продажам следующие платформы:

1. С сильным отрывом лидирует`PS4`
2. На втором месте `PS3`
3. С небольшим отставанием на третьем месте `Xone`
4. На 4-ом месте `3DS`
5. На 5-ом мете `X360`  

По остальным платформам продажи игр сильно меньше.


Определеним тренд (растут или падают) для платформ

```python
platform_trend = {}
for platform in platform_sales_sorted['Название платформы']:
    platform_data = data_actual[data_actual['platform'] == platform]
    sales_by_year = platform_data.groupby('year_of_release')['total_sales'].sum()
    years = sales_by_year.index
    sales = sales_by_year.values
    slope, _, _, _, _ = st.linregress(years, sales)
    if slope > 0:
        trend = 'Растут'
    else:
        trend = 'Падают'
    platform_trend[platform] = trend

# Вывод тренда лидирующих платформ 

print('Тренд по платформам (Растут/Падают):')
for platform,trend in platform_trend.items():
    print(platform + ': ' + trend)
```

Продемонстрируем визуально тренд по платформам (Растут/Падают)

```python
data_actual.pivot_table(
    index='year_of_release',                                                                     
    values='total_sales',
    aggfunc='sum',
    columns='platform').plot(xlim=(2013,2016),
                            figsize=(15,10),
                            xlabel='Год выпуска',
                            ylabel='Суммарные продажи (миллионы проданных копий)',
                            title='Распределение продаж игр по платформам c 2013 по 2016 год',
                            style='-o',
                            grid=True);
```

С помощью метода линейной регрсии и графика распределения продаж игр мы определили, что только по 2-ум платформам имеется общий положительный тренд продаж игр. Это `PS4 и XOne`. Эти платформы относительно молоды (3-5 лет),по всем остальным платформам общий тренд отрицательный из-за устаревания платформ.С учётом проведённого анализа интернет-магазину «Стримчик» следует сделать акцент продаж игр на платформы `PS4 и XOne`, как наиболее потенциально прибыльных.

На графике мы видим, что медианные прожажи выше всего у жанра Shooter,Sports и Platform. Что говорит нам о стабильности  жанров,котрые будут приносить доход компании.

ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков»". Для дальнейшей работы создадим новый столбец `age_category`, в котором будет расшифровка значений возрастной категории ESRB Для этого применим категоризацию.Данные по расшифровке возьмём из ['Wikipedia'](https://ru.wikipedia.org/wiki/Entertainment_Software_Rating_Board)

Узнаем  уникальные значения столбца 'rating'

Создадим функцию для категоризации значений ESRB, где укажем:
- Если категория 'M', то '17+ лет
- Если категория 'E', то '0+ лет
- Если категория 'T', то '13+ лет
- Если категория 'E10+', то '10+ лет
- Если категории нет,то 'Категория не определена'

```python
def categorize_age(rating):
    try:
        if 'M' in rating:
            return '17+ лет'
        elif 'E' in rating:
            return '0+ лет'
        elif 'T' in rating:
            return '13+ лет'
        elif 'E10+' in rating:
            return '10+ лет'
        elif 'unknown' in rating:
            return 'Категория не определена'
    except:
        pass
```

## Проверка гипотез

В рамках статистического анализа рынка компьютерных игр произведём проверку некотрых гипотез. Узнаем: 
- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые или нет?
- Средние пользовательские рейтинги жанров Action  и Sports разные или нет?

### Проверка гипотезы о cредних пользовательских рейтингов платформ Xbox One и PC 

Для проверки данной гипотезы нам понадобятся две выборки из датафрейма **data_actual**.Срез **data_actual** по значениям платформ Xbox One и PC,выборки будут сформированы по столбцу `user_score`. Данная гипотеза подходит под тип гипотезы о равенстве средних двух генеральных совокупностей. Когда сравнивают  средние двух генеральных совокупностей, нулевую гипотезу H0 можно сформулировать так: **«между средними значениями двух совокупностей нет разницы, они равны»**.Альтернативную гипотезу H1 можно сформулировать так: **«средние значения двух совокупностей не равны.»**

В наших данных имеются пропуски nan в столбце `user_score`, так как проверка гипотез с помощью статистических тестов возможна только при отсутствующих пропусках, мы отфильтруем выборки и уберём из них пропуски, оставив только заполненные значения.

```python
# фильтруем данные
df_notna = data_actual[data_actual['user_score'].notnull()]

# получаем два сэмпла
pc_score = df_notna[df_notna['platform'] == 'pc']['user_score']
xone_score = df_notna[df_notna['platform'] == 'xone']['user_score']
```

```python
alpha = 0.05 # задаём уровень статистической значимости в 5%
# Для типовой гипотезы  о равенстве средних двух генеральных совокупностей воспользуемся методом ttest_ind() 
# с указанием двух выборок, параметра equal_var=False(разные размеры выборок) 
# и параметра alternative='two-sided'(средние значения двух совокупностей не равны.)  

results = st.ttest_ind(pc_score, 
                       xone_score,
                       equal_var=False,
                       alternative='two-sided'
                      )   
```

```python
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Полученное значение p-value (~ 0.1475959) больше заданного уровня статистической значимости, поэтому  нулевую гипотезу не отвергаем. Есть основания говорить, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

### Проверка гипотезы о средних пользовательских рейтингов жанра Action и Sports

Для проверки данной гипотезы нам понадобятся две выборки из датафрейма **data_actual**,срез **data_actual** по значениям жанров Action и Sports,выборки будут сформированы по столбцу `user_score`. Данная гипотеза подходит под тип гипотезы о равенстве средних двух генеральных совокупностей. Когда сравнивают  средние двух генеральных совокупностей, нулевую гипотезу H0 можно сформулировать так: **«между средними значениями двух совокупностей нет разницы, они равны»**.Альтернативную гипотезу H1 можно сформулировать так: **«средние значения двух совокупностей не равны.»**

Создадим 2-е переменные и добавим в них отфильтрованные значения со значением столбца 'user_score'

```python
# получаем два сэмпла
action_score = df_notna[df_notna['genre'] == 'Action']['user_score']
sports_score = df_notna[df_notna['genre'] == 'Sports']['user_score']
```

```python
alpha = 0.05 # задаём уровень статистической значимости в 5%
# Для типовой гипотезы  о равенстве средних двух генеральных совокупностей воспользуемся методом ttest_ind() 
# с указанием двух выборок, параметра equal_var=False(разные размеры выборок) 
# и параметра alternative='two-sided'(средние значения двух совокупностей не равны.)  

results = st.ttest_ind(action_score, 
                       sports_score,
                       equal_var=False,
                       alternative='two-sided'
                      )   
```

```python
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Полученное значение p-value (~ 1.4460039700704318e-20) меньше заданного уровня значимости, поэтому можно отвергнуть нулевую гипотезу. Есть основания говорить, что cредние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.
