**Описание проекта**

В рамках данного исследования мы произведём анализ популярного сервиса аренды самокатов GoFast. Нам передали данные о некоторых пользователях из нескольких городов, а также об их поездках. 

**Описание тарифов**

Нам известно,что для совершения поездки по городу, пользователи сервиса GoFast пользуются мобильным приложением. Сервисом можно пользоваться:
- без подписки
  - абонентская плата отсутствует;
  - стоимость одной минуты поездки — 8 рублей;
  - стоимость старта (начала поездки) — 50 рублей;
- с подпиской Ultra
  - абонентская плата — 199 рублей в месяц;
  - стоимость одной минуты поездки — 6 рублей;
  - стоимость старта — бесплатно.

**Предоставленные файлы**

- Информация о пользователях — `users_go.csv`
- Информация о поездках — `rides_go.csv`
- Информация о подписках — `subscriptions_go.csv`

**Цель исследования**

Проанализировать данные и проверить некоторые гипотезы, которые могут помочь бизнесу вырасти.В рамках данного исследования ответим на следующие вопросы:
- Тратят ли пользователи с подпиской больше времени на поездки?
- Можно ли сказать, что расстояние, которое проезжают пользователи с подпиской за одну поездку, не превышает 3130 метров? 
- Будет ли помесячная выручка от пользователей с подпиской по месяцам выше, чем выручка от пользователей без подписки?

Рассчитать количество промокодов рассылки для выполнения плана.

Рассчитать вероятность того, что уведомление откроют не более 399,5 тыс. пользователей

## Объединение данных

## Проверка гипотез

Продакт-менеджеры сервиса хотят увеличить количество пользователей с подпиской. Для этого они будут проводить различные акции, но сначала нужно выяснить несколько важных моментов. В этом нам помогут гипотезы.

### Проверка гипотезы о затратах времени на поездки пользователей с подпиской и без подписки

Важно понять, тратят ли пользователи с подпиской больше времени на поездки? Если да, то пользователи с подпиской могут быть «выгоднее» для компании. Проверим гипотезу. Используем исходные данные о продолжительности каждой сессии — отдельно для подписчиков и тех, у кого нет подписки.

Для проверки данной гипотезы нам понадобятся две выборки из датафрейма **data_free** и **data_ultra**.Данная гипотеза подходит под тип гипотезы о равенстве средних двух генеральных совокупностей. Когда сравнивают  средние двух генеральных совокупностей, нулевую гипотезу можно сформулировать так: **«средние значения совокупностей равны»**.Альтернативную гипотезу можно сформулировать так: **«Среднее значение первой генеральной совокупности больше среднего значения второй генеральной совокупности.»**

```python
alpha = 0.05 # задаём уровень статистической значимости в 5%
# Для типовой гипотезы  о равенстве средних двух генеральных совокупностей воспользуемся методом ttest_ind() 
# с указанием двух выборок, параметра equal_var=False(разные размеры выборок) 
# и параметра alternative='greater'(Среднее значение первой генеральной совокупности 
# больше среднего значения второй генеральной совокупности.)  
results = st.ttest_ind(data_ultra['duration'], 
                       data_free['duration'],
                       equal_var=False,
                       alternative='greater'
                      )   
```

```python
print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Полученное значение p-value (~ 5.6757813771289775e-37) меньше заданного уровня значимости, поэтому можно отвергнуть нулевую гипотезу. Есть основания говорить, что пользователи с подпиской тратят больше времени на поездки.

### Проверка гипотезы о преодоленном расстоянии пользователей за одну поездку

Расстояние одной поездки в 3130 метров — оптимальное с точки зрения износа самоката. Можно ли сказать, что расстояние, которое проезжают пользователи с подпиской за одну поездку, не превышает 3130 метров? Проверим гипотезу и сделаем выводы.

Для проверки данной гипотезы нам понадобится генеральная совокупность из датафрейма **data_ultra**.Данная гипотеза подходит под тип гипотезы об **"Проверке односторонних гипотез в Python"** равно ли истинное среднее генеральной совокупности заданному числу. Когда сравнивают среднее  генеральной совокупности с заданным числом, нулевую гипотезу можно сформулировать так: **«среднее значение генеральной совокупности равно заданному числу»**.Альтернативную гипотезу можно сформулировать так: **«среднее значение генеральной совокупности меньше заданного числа.»**

```python
alpha = 0.05 # задаём уровень статистической значимости в 5%
prev_screens_value = 3130 # заданное число
 
# Для типовой гипотезы  о равенстве истинного среднего генеральной совокупности заданному числу воспользуемся 
# воспользуемся методом ttest_1samp() с указанием генеральной совокупности,заданного числа и 
# параметра alternative='less'(Среднее значение  генеральной совокупности меньше заданного числа).)
results_2 = st.ttest_1samp(data_ultra['distance'],prev_screens_value,alternative='less')
```

```python
print('p-значение:', results_2.pvalue)

if results_2.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Полученное значение p-value (~ 0.080463) больше заданного уровня значимости, поэтому  нулевую гипотезу не отвергаем. Есть основания говорить, что пользователи с подпиской проезжают около 3130 метров.

### Гипотеза о равенстве средних для зависимых (парных) выборок

Представим такую ситуацию: техническая команда сервиса обновила сервера, с которыми взаимодействует мобильное приложение. Она надеется, что из-за этого количество обращений в техподдержку значимо снизилось. Некоторый файл содержит для каждого пользователя данные о количестве обращений до обновления и после него. Определим какой тест нам понадобился бы для проверки этой гипотезы.

Для проверки данной гипотезы нам понадобится **Гипотеза о равенстве средних для зависимых (парных) выборок**.Мы анализируем одну генеральную совокупность, с которой произошли некоторые изменения. В этом случае мы будем проверять, равны ли средние совокупности до и после этого изменения.Гипотезы в таком случае будут сформулированы так:
H0: u1=u2
H1: u1!=u2
где,u1 и u2 — средние значения до и после изменения. Чтобы проверить гипотезу о равенстве средних двух генеральных совокупностей для зависимых (парных) выборок в Python, применим метод scipy.stats.ttest_rel().Ему достаточно передать всего два параметра: наборы данных до и после. Оба набора должны быть одинаковыми по размеру: каждое измерение должно быть проведено до и после изменения.

**Вывод:** В рамках статистического анализа, мы проверили некоторые гипотезы и получили следующий результат:
- пользователи с подпиской тратят больше времени на поездки.
- пользователи с подпиской проезжают около 3130 метров за одну поездку
- помесячная выручка пользователей с подпиской и без подписки существенно не отличается

### расчёт количества промокодов

Отделу маркетинга GoFast поставили задачу: нужно провести акцию с раздачей промокодов на один бесплатный месяц подписки, в рамках которой как минимум 100 существующих клиентов должны продлить эту подписку. То есть по завершении периода действия подписки пользователь может либо отказаться от неё, либо продлить, совершив соотвествующий платёж. Эта акция уже проводилась ранее и по итогу выяснилось, что после бесплатного пробного периода подписку продлевают 10 % пользователей. Выясним, какое минимальное количество промокодов нужно разослать, чтобы вероятность не выполнить план была примерно 5 %. Подберём параметры распределения, описывающего эту ситуацию, построим график распределения и сформулируем ответ на вопрос о количестве промокодов.

Ситуация представлена биномиальным экспериментом (серия независимых экспериментов по схеме Бернулли). Для указанной ситуации построим график биномиального распределения, который описывает указанную ситуацию.

```python
from scipy.stats import binom
```

Количество испытаний(n) легко получить зная его математическое ожидание E(X)=100 и вероятность успеха p=0.1 n = E(X)*P 

```python
n = 1000 # количество испытаний
p = 0.1 # вероятность успеха в каждом из них

plt.figure(figsize=(8,8))

distr_0 = []# список distr, в котором будем хранить значения распределения вероятностей


# применим метод binom.pmf() из модуля scipy.stats. Он помогает получить вероятности биномиального распределения
for k in range(0,n+1):    
    current_value = binom.pmf(k,n,p) 
    distr_0.append(current_value)
    

plt.bar(range(0,n+1),distr_0); # построим распределение вероятностей
plt.xlabel('Количество испытаний (ед.)');
plt.ylabel('Вероятность успеха');
plt.show();
```

По графику биномиального распределения видно, что самое вероятное значение это 100.Однако не понятно какая вероятность получить не менее 100 подписок в 2 ой и следующие месяцы (задача найти вероятность «хвоста» распределения)при количестве рассылок равным 1000. Для этого нам понадобится нормальное распределение и метод cdf(получает на вход один аргумент и возвращает вероятность, что случайная величина примет значение, меньшее заданного аргумента).

```python
from math import sqrt

# задаём параметры биномиального распределения

n = 1000
p = 0.1

#зададим параметры нормального распределения равными мат.ожиданию и дисперсии биномиального распределения

mu = n * p
sigma = sqrt(n * p * (1 - p))

# задаём нормальное распределение

distr_1 = st.norm(mu, sigma)

# считаем вероятность получить значение

result=  distr_1.cdf(100)
print(result)
```

Вероятность получить 100 или меньше подписок в 2 ой и следующие месяцы составляет 50%. От нас требуется чтобы вероятность не выполнить план была примерно 5%, тоесть мы должны подобрать подходящее значение количества испытаний(n) для заданного параметра p=0.1 

Эспериментальным путём мы подберём такое значение количества испытаний(n) для заданного параметра p=0.1, чтобы вероятность в результате получить 100 или меньше подписок составляла ~ 5%

```python
from math import sqrt

# задаём параметры биномиального распределения
n = 1168
p = 0.1

#зададим параметры нормального распределения равными мат.ожиданию и дисперсии биномиального распределения

mu = n * p
sigma = sqrt(n * p * (1 - p))

# задаём нормальное распределение

distr_2 = st.norm(mu, sigma)

# считаем вероятность получить значение

result=  distr_2.cdf(100)
print(result)
```

По результатам расчётов мы получили, что мимумум **1168 промокодов** нужно разослать для того,чтобы вероятность не выполнить план была примерно 5 % при 100 подписках в 2 ой и следующие месяцы.

Для этого нам понадобится нормальное распределение и метод cdf(получает на вход один аргумент и возвращает вероятность, что случайная величина примет значение, меньшее заданного аргумента).

```python
import scipy.stats as st
from math import sqrt
# задаём параметры биномиального распределения

n = 1000000
p = 0.4

#зададим параметры нормального распределения равными мат.ожиданию и дисперсии биномиального распределения

mu = n * p
sigma = sqrt(n * p * (1 - p))

# задаём нормальное распределение

distr_4 = st.norm(mu, sigma)

# считаем вероятность получить значение правого хвоста

result=   distr_4.cdf(399500)
print(result)
```
